```python
import pandas as pd
import os

file_path = 'train/train.csv' 
data = pd.read_csv(file_path)
data.loc[(data['품종명'] == '후지') & (data['등급'] == '상품') & (data['거래단위'] == '10 개'), '품종명'] = '후지홍로'
data.loc[(data['품종명'] == '홍로') & (data['등급'] == '상품') & (data['거래단위'] == '10 개'), '품종명'] = '후지홍로'

def create_shifted_features(df, column_name, num_shifts=8):
    for i in range(1, num_shifts + 1):
        df[f'{column_name}_T-{i}'] = df[column_name].shift(i)
    return df

def create_future_features(df, column_name, num_future_shifts=3):
    for i in range(1, num_future_shifts + 1):
        df[f'{column_name}_{i}순'] = df[column_name].shift(-i)
    return df

train_df = data.groupby(['품목명', '품종명', '거래단위', '등급']).apply(lambda x: create_shifted_features(x, '평년 평균가격(원)')).reset_index(drop=True)
train_df = train_df.groupby(['품목명', '품종명', '거래단위', '등급']).apply(lambda x: create_shifted_features(x, '평균가격(원)')).reset_index(drop=True)
train_df = train_df.groupby(['품목명', '품종명', '거래단위', '등급']).apply(lambda x: create_future_features(x, '평균가격(원)')).reset_index(drop=True)
train_df = train_df.dropna()
train_df.rename(columns={'평균가격(원)_1순': '1순', '평균가격(원)_2순': '2순', '평균가격(원)_3순' : '3순'}, inplace=True)
train_df.to_csv('Processtr.csv',index=False,encoding='cp949')

def process_file(file_path, file_name):
    data = pd.read_csv(file_path)
    data.loc[(data['품종명'] == '후지') & (data['등급'] == '상품') & (data['거래단위'] == '10 개'), '품종명'] = '후지홍로'
    data.loc[(data['품종명'] == '홍로') & (data['등급'] == '상품') & (data['거래단위'] == '10 개'), '품종명'] = '후지홍로'
    def create_shifted_features(df, column_name, num_shifts=8):
        for i in range(1, num_shifts + 1):
            df[f'{column_name}_T-{i}'] = df[column_name].shift(i)
        return df
    data = data.groupby(['품목명', '품종명', '거래단위', '등급']).apply(lambda x: create_shifted_features(x, '평년 평균가격(원)')).reset_index(drop=True)
    data = data.groupby(['품목명', '품종명', '거래단위', '등급']).apply(lambda x: create_shifted_features(x, '평균가격(원)')).reset_index(drop=True)
    data = data[data['시점'] == 'T']
    data['1순'] = 0
    data['2순'] = 0
    data['3순'] = 0
    data['시점'] = file_name.replace('.csv', '') 
    return data
all_data = [] 
for i in range(25):
    file_name = f'TEST_{i:02d}.csv' 
    file_path = os.path.join('test', file_name)
    processed_data = process_file(file_path, file_name)
    all_data.append(processed_data) 

final_data = pd.concat(all_data, ignore_index=True)

final_data_gochu = final_data[(final_data['품목명'] == '건고추') & (final_data['품종명'] == '화건') & (final_data['거래단위'] == '30 kg') & (final_data['등급'] == '상품')]
final_data_gochu = final_data_gochu.reset_index(drop = True)

final_data_apple = final_data[(final_data['품목명'] == '사과') & (final_data['품종명'] == '후지홍로') & (final_data['거래단위'] == '10 개') & (final_data['등급'] == '상품')]
final_data_apple = final_data_apple.reset_index(drop = True)

final_data_potato = final_data[(final_data['품목명'] == '감자') & (final_data['품종명'] == '감자 수미') & (final_data['거래단위'] == '20키로상자') & (final_data['등급'] == '상')]
final_data_potato = final_data_potato.reset_index(drop = True)

final_data_pear = final_data[(final_data['품목명'] == '배') & (final_data['품종명'] == '신고') & (final_data['거래단위'] == '10 개') & (final_data['등급'] == '상품')]
final_data_pear = final_data_pear.reset_index(drop = True)

final_data_garlic = final_data[(final_data['품목명'] == '깐마늘(국산)') & (final_data['품종명'] == '깐마늘(국산)') & (final_data['거래단위'] == '20 kg') & (final_data['등급'] == '상품')]
final_data_garlic = final_data_garlic.reset_index(drop = True)

final_data_moo = final_data[(final_data['품목명'] == '무') & (final_data['품종명'] == '무') & (final_data['거래단위'] == '20키로상자') & (final_data['등급'] == '상')]
final_data_moo = final_data_moo.reset_index(drop = True)

final_data_lettuce = final_data[(final_data['품목명'] == '상추') & (final_data['품종명'] == '청') & (final_data['거래단위'] == '100 g') & (final_data['등급'] == '상품')]
final_data_lettuce = final_data_lettuce.reset_index(drop = True)

final_data_cabbage = final_data[(final_data['품목명'] == '배추') & (final_data['품종명'] == '배추') & (final_data['거래단위'] == '10키로망대') & (final_data['등급'] == '상')]
final_data_cabbage = final_data_cabbage.reset_index(drop = True)

final_data_onion = final_data[(final_data['품목명'] == '양파') & (final_data['품종명'] == '양파') & (final_data['거래단위'] == '1키로') & (final_data['등급'] == '상')]
final_data_onion = final_data_onion.reset_index(drop = True)

final_data_daepa = final_data[(final_data['품목명'] == '대파') & (final_data['품종명'] == '대파(일반)') & (final_data['거래단위'] == '1키로단') & (final_data['등급'] == '상')]
final_data_daepa = final_data_daepa.reset_index(drop = True)

Filter_final_data = pd.concat([final_data_gochu, final_data_apple, final_data_potato, final_data_pear, final_data_garlic, final_data_moo, final_data_lettuce, final_data_cabbage, final_data_onion, final_data_daepa],axis=0)
Filter_final_data = Filter_final_data.reset_index(drop = True)
Filter_final_data.to_csv('Processte.csv',index=False, encoding='CP949')
def process_file(file_path, file_name):
    data = pd.read_csv(file_path)
    data.loc[(data['품종명'] == '후지') & (data['등급'] == '상품') & (data['거래단위'] == '10 개'), '품종명'] = '후지홍로'
    data.loc[(data['품종명'] == '홍로') & (data['등급'] == '상품') & (data['거래단위'] == '10 개'), '품종명'] = '후지홍로'
    def create_shifted_features(df, column_name, num_shifts=8):
        for i in range(1, num_shifts + 1):
            df[f'{column_name}_T-{i}'] = df[column_name].shift(i)
        return df
    data = data.groupby(['품목명', '품종명', '거래단위', '등급']).apply(lambda x: create_shifted_features(x, '평년 평균가격(원)')).reset_index(drop=True)
    data = data.groupby(['품목명', '품종명', '거래단위', '등급']).apply(lambda x: create_shifted_features(x, '평균가격(원)')).reset_index(drop=True)
    data = data[data['시점'] == 'T']
    data['1순'] = 0
    data['2순'] = 0
    data['3순'] = 0
    data['시점'] = file_name.replace('.csv', '')  
    return data

all_data = []  

for i in range(25): 
    file_name = f'TEST_{i:02d}.csv'  
    file_path = os.path.join('test', file_name)  
    processed_data = process_file(file_path, file_name) 
    all_data.append(processed_data)  

t_df = pd.concat(all_data, ignore_index=True)

data = {
    '품목명': ['건고추', '사과', '감자', '배', '깐마늘(국산)', '무', '상추', '배추', '양파', '대파'],
    '품종명': ['화건', '후지홍로', '감자 수미', '신고', '깐마늘(국산)', '무', '청', '배추', '양파', '대파(일반)'],
    '거래단위': ['30 kg', '10 개', '20키로상자', '10 개', '20 kg', '20키로상자', '100 g', '10키로망대', '1키로', '1키로단'],
    '등급': ['상품', '상품', '상', '상품', '상품', '상', '상품', '상', '상', '상']
}
df = pd.DataFrame(data)
transformed_item_dfs = {}
for _, criteria in df.iterrows():
    item = criteria['품목명']
    main_variety = criteria['품종명']
    main_unit = criteria['거래단위']
    main_grade = criteria['등급']
    main_variety_df = train_df[
        (train_df['품목명'] == item) &
        (train_df['품종명'] == main_variety) &
        (train_df['거래단위'] == main_unit) &
        (train_df['등급'] == main_grade)
    ].reset_index(drop=True)
    other_varieties_df = train_df[
        (train_df['품목명'] == item) &
        ~(
            (train_df['품종명'] == main_variety) &
            (train_df['거래단위'] == main_unit) &
            (train_df['등급'] == main_grade)
        )
    ].reset_index(drop=True)
    pivoted_varieties_df = other_varieties_df.pivot_table(
        index='시점',
        columns=['품종명', '거래단위', '등급'],
        values=[col for col in other_varieties_df.columns if col not in ['시점', '품목명', '품종명', '거래단위', '등급']]
    ).reset_index()
    pivoted_varieties_df.columns = ['_'.join([str(c) for c in col if c]).strip() if isinstance(col, tuple) else col for col in pivoted_varieties_df.columns]
    merged_df = pd.merge(main_variety_df, pivoted_varieties_df, on='시점', how='left')
    if item not in transformed_item_dfs:
        transformed_item_dfs[item] = []
    transformed_item_dfs[item].append(merged_df)


final_item_dfs = {}
for item, dfs in transformed_item_dfs.items():
    final_item_dfs[item] = pd.concat(dfs, ignore_index=True)


df_건고추 = final_item_dfs.get('건고추', pd.DataFrame())
df_사과 = final_item_dfs.get('사과', pd.DataFrame())
df_감자 = final_item_dfs.get('감자', pd.DataFrame())
df_배 = final_item_dfs.get('배', pd.DataFrame())
df_깐마늘_국산 = final_item_dfs.get('깐마늘(국산)', pd.DataFrame())
df_무 = final_item_dfs.get('무', pd.DataFrame())
df_상추 = final_item_dfs.get('상추', pd.DataFrame())
df_배추 = final_item_dfs.get('배추', pd.DataFrame())
df_양파 = final_item_dfs.get('양파', pd.DataFrame())
df_대파 = final_item_dfs.get('대파', pd.DataFrame())
columns_to_keep = ['1순', '2순', '3순']
columns_to_exclude = df_건고추.columns[df_건고추.columns.str.contains('1순|2순|3순') & ~df_건고추.columns.isin(columns_to_keep)]
df_건고추 = df_건고추.drop(columns=columns_to_exclude)

columns_to_exclude = df_사과.columns[df_사과.columns.str.contains('1순|2순|3순') & ~df_사과.columns.isin(columns_to_keep)]
df_사과 = df_사과.drop(columns=columns_to_exclude)

columns_to_exclude = df_감자.columns[df_감자.columns.str.contains('1순|2순|3순') & ~df_감자.columns.isin(columns_to_keep)]
df_감자 = df_감자.drop(columns=columns_to_exclude)

columns_to_exclude = df_배.columns[df_배.columns.str.contains('1순|2순|3순') & ~df_배.columns.isin(columns_to_keep)]
df_배 = df_배.drop(columns=columns_to_exclude)

columns_to_exclude = df_깐마늘_국산.columns[df_깐마늘_국산.columns.str.contains('1순|2순|3순') & ~df_깐마늘_국산.columns.isin(columns_to_keep)]
df_깐마늘_국산 = df_깐마늘_국산.drop(columns=columns_to_exclude)

columns_to_exclude = df_무.columns[df_무.columns.str.contains('1순|2순|3순') & ~df_무.columns.isin(columns_to_keep)]
df_무 = df_무.drop(columns=columns_to_exclude)

columns_to_exclude = df_상추.columns[df_상추.columns.str.contains('1순|2순|3순') & ~df_상추.columns.isin(columns_to_keep)]
df_상추 = df_상추.drop(columns=columns_to_exclude)

columns_to_exclude = df_배추.columns[df_배추.columns.str.contains('1순|2순|3순') & ~df_배추.columns.isin(columns_to_keep)]
df_배추 = df_배추.drop(columns=columns_to_exclude)

columns_to_exclude = df_양파.columns[df_양파.columns.str.contains('1순|2순|3순') & ~df_양파.columns.isin(columns_to_keep)]
df_양파 = df_양파.drop(columns=columns_to_exclude)

columns_to_exclude = df_대파.columns[df_대파.columns.str.contains('1순|2순|3순') & ~df_대파.columns.isin(columns_to_keep)]
df_대파 = df_대파.drop(columns=columns_to_exclude)

df_건고추.to_csv('train_건고추.csv',index=False,encoding='cp949')

df_사과.to_csv('train_사과.csv',index=False,encoding='cp949')

df_감자.to_csv('train_감자.csv',index=False,encoding='cp949')

df_배.to_csv('train_배.csv',index=False,encoding='cp949')

df_깐마늘_국산.to_csv('train_깐마늘_국산.csv',index=False,encoding='cp949')

df_무.to_csv('train_무.csv',index=False,encoding='cp949')

df_상추.to_csv('train_상추.csv',index=False,encoding='cp949')

df_배추.to_csv('train_배추.csv',index=False,encoding='cp949')

df_양파.to_csv('train_양파.csv',index=False,encoding='cp949')

df_대파.to_csv('train_대파.csv',index=False,encoding='cp949')

data = {
    '품목명': ['건고추', '사과', '감자', '배', '깐마늘(국산)', '무', '상추', '배추', '양파', '대파'],
    '품종명': ['화건', '후지홍로', '감자 수미', '신고', '깐마늘(국산)', '무', '청', '배추', '양파', '대파(일반)'],
    '거래단위': ['30 kg', '10 개', '20키로상자', '10 개', '20 kg', '20키로상자', '100 g', '10키로망대', '1키로', '1키로단'],
    '등급': ['상품', '상품', '상', '상품', '상품', '상', '상품', '상', '상', '상']
}

df = pd.DataFrame(data)
transformed_item_dfs = {}
for _, criteria in df.iterrows():
    item = criteria['품목명']
    main_variety = criteria['품종명']
    main_unit = criteria['거래단위']
    main_grade = criteria['등급']
    main_variety_df = t_df[
        (t_df['품목명'] == item) &
        (t_df['품종명'] == main_variety) &
        (t_df['거래단위'] == main_unit) &
        (t_df['등급'] == main_grade)
    ].reset_index(drop=True)
    other_varieties_df = t_df[
        (t_df['품목명'] == item) &
        ~(
            (t_df['품종명'] == main_variety) &
            (t_df['거래단위'] == main_unit) &
            (t_df['등급'] == main_grade)
        )
    ].reset_index(drop=True)
    pivoted_varieties_df = other_varieties_df.pivot_table(
        index='시점',
        columns=['품종명', '거래단위', '등급'],
        values=[col for col in other_varieties_df.columns if col not in ['시점', '품목명', '품종명', '거래단위', '등급']]
    ).reset_index()
    pivoted_varieties_df.columns = ['_'.join([str(c) for c in col if c]).strip() if isinstance(col, tuple) else col for col in pivoted_varieties_df.columns]
    merged_df = pd.merge(main_variety_df, pivoted_varieties_df, on='시점', how='left')
    if item not in transformed_item_dfs:
        transformed_item_dfs[item] = []
    transformed_item_dfs[item].append(merged_df)


final_item_dfs = {}
for item, dfs in transformed_item_dfs.items():
    final_item_dfs[item] = pd.concat(dfs, ignore_index=True)

df_건고추 = final_item_dfs.get('건고추', pd.DataFrame())
df_사과 = final_item_dfs.get('사과', pd.DataFrame())
df_감자 = final_item_dfs.get('감자', pd.DataFrame())
df_배 = final_item_dfs.get('배', pd.DataFrame())
df_깐마늘_국산 = final_item_dfs.get('깐마늘(국산)', pd.DataFrame())
df_무 = final_item_dfs.get('무', pd.DataFrame())
df_상추 = final_item_dfs.get('상추', pd.DataFrame())
df_배추 = final_item_dfs.get('배추', pd.DataFrame())
df_양파 = final_item_dfs.get('양파', pd.DataFrame())
df_대파 = final_item_dfs.get('대파', pd.DataFrame())

columns_to_keep = ['1순', '2순', '3순']
columns_to_exclude = df_건고추.columns[df_건고추.columns.str.contains('1순|2순|3순') & ~df_건고추.columns.isin(columns_to_keep)]
df_건고추 = df_건고추.drop(columns=columns_to_exclude)

columns_to_exclude = df_사과.columns[df_사과.columns.str.contains('1순|2순|3순') & ~df_사과.columns.isin(columns_to_keep)]
df_사과 = df_사과.drop(columns=columns_to_exclude)

columns_to_exclude = df_감자.columns[df_감자.columns.str.contains('1순|2순|3순') & ~df_감자.columns.isin(columns_to_keep)]
df_감자 = df_감자.drop(columns=columns_to_exclude)

columns_to_exclude = df_배.columns[df_배.columns.str.contains('1순|2순|3순') & ~df_배.columns.isin(columns_to_keep)]
df_배 = df_배.drop(columns=columns_to_exclude)

columns_to_exclude = df_깐마늘_국산.columns[df_깐마늘_국산.columns.str.contains('1순|2순|3순') & ~df_깐마늘_국산.columns.isin(columns_to_keep)]
df_깐마늘_국산 = df_깐마늘_국산.drop(columns=columns_to_exclude)

columns_to_exclude = df_무.columns[df_무.columns.str.contains('1순|2순|3순') & ~df_무.columns.isin(columns_to_keep)]
df_무 = df_무.drop(columns=columns_to_exclude)

columns_to_exclude = df_상추.columns[df_상추.columns.str.contains('1순|2순|3순') & ~df_상추.columns.isin(columns_to_keep)]
df_상추 = df_상추.drop(columns=columns_to_exclude)

columns_to_exclude = df_배추.columns[df_배추.columns.str.contains('1순|2순|3순') & ~df_배추.columns.isin(columns_to_keep)]
df_배추 = df_배추.drop(columns=columns_to_exclude)

columns_to_exclude = df_양파.columns[df_양파.columns.str.contains('1순|2순|3순') & ~df_양파.columns.isin(columns_to_keep)]
df_양파 = df_양파.drop(columns=columns_to_exclude)

columns_to_exclude = df_대파.columns[df_대파.columns.str.contains('1순|2순|3순') & ~df_대파.columns.isin(columns_to_keep)]
df_대파 = df_대파.drop(columns=columns_to_exclude)


df_건고추.to_csv('test_건고추.csv',index=False,encoding='cp949')

df_사과.to_csv('test_사과.csv',index=False,encoding='cp949')

df_감자.to_csv('test_감자.csv',index=False,encoding='cp949')

df_배.to_csv('test_배.csv',index=False,encoding='cp949')

df_깐마늘_국산.to_csv('test_깐마늘_국산.csv',index=False,encoding='cp949')

df_무.to_csv('test_무.csv',index=False,encoding='cp949')

df_상추.to_csv('test_상추.csv',index=False,encoding='cp949')

df_배추.to_csv('test_배추.csv',index=False,encoding='cp949')

df_양파.to_csv('test_양파.csv',index=False,encoding='cp949')

df_대파.to_csv('test_대파.csv',index=False,encoding='cp949')
```
